```python
data_DE=np.array([[4.79,4.93,5.05,5.11],[4.79,4.86,4.90,5.01,4.93,5.13],[4.52,4.93,4.79,4.86,4.93]],dtype=object)
   
  
    

n1=np.array([2,3,4,5])
n2=np.array([2,3,4,5,6,7])
n3=np.array([2,3,4,5,6])



def linear_function(X, a, b):
    return a * X + b


fig, (ax1,ax2) = plt.subplots(2,1,figsize=(6,8),gridspec_kw={'height_ratios': [2, 1]}) 



popt1, pcov1 = curve_fit(linear_function, n1,data_DE[0])
popt2, pcov2 = curve_fit(linear_function, n2,data_DE[1])

popt3, pcov3 = curve_fit(linear_function, n3,data_DE[2])


# Get the slope and intercept
slope1 = popt1[0]
intercept1 = popt1[1]


slope2 = popt2[0]
intercept2 = popt2[1]

slope3 = popt3[0]
intercept3 = popt3[1]

# Get the standard errors for the parameters
slope_std_err1 = np.sqrt(pcov1[0, 0])
intercept_std_err1 = np.sqrt(pcov1[1, 1])

slope_std_err2 = np.sqrt(pcov2[0, 0])
intercept_std_err2 = np.sqrt(pcov2[1, 1])

slope_std_err3 = np.sqrt(pcov3[0, 0])
intercept_std_err3 = np.sqrt(pcov3[1, 1])



Y_pred1 = linear_function(n1, *popt1)
Y_pred2 = linear_function(n2, *popt2)
Y_pred3 = linear_function(n3, *popt3)


residuals1 = data_DE[0] - Y_pred1

residuals2 = data_DE[1] - Y_pred2

residuals3 = data_DE[2] - Y_pred3
#residuals = y - Y_pred
ax1.plot(n1, Y_pred1, ls="--", linewidth=2,c="k")
ax1.plot(n2, Y_pred2, ls="--", linewidth=2,c="b")
ax1.plot(n3, Y_pred3, ls="--", linewidth=2,c="r")

ax1.errorbar(n1,data_DE[0],yerr=[0.01]*len(n1),fmt="o",label="T=190Â°C",c="k")


ax1.errorbar(n2,data_DE[1],yerr=[0.01]*len(n2),fmt="o",label="T=214Â°C",c="b")


ax1.errorbar(n3,data_DE[2],yerr=[0.01]*len(n3),fmt="o",label="T=200Â°C",c="r")


ax1.set_xlabel('Orden del minimo')
ax1.set_ylabel('Energia entre los minimos(eV)')
ax1.legend(loc='upper right')
ax1.set_title(f'$\Delta$ E de los minimos')


ax2.scatter(n1,residuals1,label="T=190Â°C",c="k")
ax2.scatter(n2,residuals2,label="T=214Â°C",c="b")
ax2.scatter(n3,residuals3,label="T=200Â°C",c="r")
ax2.axhline(y=0,color='black', linewidth=1, linestyle='--')

ax2.set_title("Residuales")


plt.tight_layout()



ax1.grid()
ax2.grid()


plt.savefig('Delta_E.pdf', format='pdf')
plt.show()
```
